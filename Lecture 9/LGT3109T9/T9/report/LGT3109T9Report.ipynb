### Part 3: Operations on ZIP Files

#### Write and execute codes for the following questions 1-4. (32 points)

Question 1. Import module zipfile. Print file names included in data12.zip in data folder of your working directory, and print their file sizes.

```python
#Write your code here:
import zipfile
from pathlib import Path

zip_path = Path.cwd() / 'data' / 'data12.zip'

with zipfile.ZipFile(zip_path, 'r') as zf:
    for info in zf.infolist():
        print(f'{info.filename}: {info.file_size} bytes')
```

Question 2. Create a folder named 'output' in your working folder if it does not exist. Extract all files contained in data12.zip to 'output'. Print all files in the folder 'output'

```python
#Write your code here:
output_folder = Path.cwd() / 'output'
output_folder.mkdir(exist_ok=True)

zip_path = Path.cwd() / 'data' / 'data12.zip'
with zipfile.ZipFile(zip_path, 'r') as zf:
    zf.extractall(output_folder)

for file in output_folder.iterdir():
    print(file.name)
```

Question 3. Print and extract each file contained in data23.zip to the folder 'output' if the file does not exist in 'output'. Print all files in folder 'output'.

```python
#Write your code here:
output_folder = Path.cwd() / 'output'
zip_path = Path.cwd() / 'data' / 'data23.zip'

with zipfile.ZipFile(zip_path, 'r') as zf:
    for info in zf.infolist():
        print(info.filename)
        out_file = output_folder / info.filename
        if not out_file.exists():
            zf.extract(info.filename, output_folder)

for file in output_folder.iterdir():
    print(file.name)
```

Question 4. Compress all files in output to a zip file named output.zip in your working directory. Print the content of output.zip

```python
#Write your code here:
output_folder = Path.cwd() / 'output'
zip_file = Path.cwd() / 'output.zip'

with zipfile.ZipFile(zip_file, 'w') as zf:
    for file in output_folder.iterdir():
        zf.write(file, arcname=file.name)

with zipfile.ZipFile(zip_file, 'r') as zf:
    for info in zf.infolist():
        print(info.filename)
```
