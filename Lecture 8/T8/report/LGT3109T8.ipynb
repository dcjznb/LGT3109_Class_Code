#### 2.1. Password Checker

Write a program to prompt for a user to input a line of password, and to print 'pass' if the password contains only lower case letters, but contains no letter that appears more than three times, and print 'fail' otherwise

Sample input/output:

    Enter a password: abcABC

##### Question 1: Explain how you are going to define a dictionary variable ``char_count`` that represents the number of times that each letter appears in the input password? What are its keys and value? (3 points)

<font color='red'>Edit this cell to answer the question above:</font>  
The keys of the dictionary variable char_count are the letters in the input password, and the values are the number of times that each letter appears.



##### Question 2: Explain how you are going to define a boolean variable ``flag_fail`` that indicates whether the input string contains a characters that is not a lower case letter, or contains a letter that appears more than three times. What's its initial value? (3 points)

<font color='red'>Edit this cell to answer the question above:</font>  
flag_fail is initialized as False. When we find a character that is not a lower case letter or a letter that appears more than three times, we set flag_fail to True.



##### Question 3: Explain how you are going to use a for loop to traverse each character ``c`` appearing in the input password. For each character ``c`` traversed, how are you going to update the dictionary variable ``char_count`` and how are you going to update the boolean variable ``flg_fail``? (3 points)

<font color='red'>Edit this cell to answer the question above:</font>  
For each c in the input password, if c is not a lower case letter or char_count[c] > 3, we set flag_fail to True. Otherwise, we increment char_count[c] by 1.


##### Question 4: Draw a flow chart to explain how the code will work. (3 points)

##### Question 5: Write the code: (10 points)

```python
#edit this cell to write your code
password = input('Enter a password:')
char_count = {}
flag_fail = False
for c in password:
    if not c.islower():
        flag_fail = True
        break
    char_count[c] = char_count.get(c, 0) + 1
    if char_count[c] > 3:
        flag_fail = True
        break

if flag_fail:
    print('fail')
else:
    print('pass')
```

### Part 5 Tuples: Comparing

#### 5.1. Read the code below and answer the question (10 points):

```python
sales = {'Hong Kong':90, 'Guangzhou':98, 'Tokyo': 88, 'San Francisco': 99}
d = {}
for (city,sale) in sales.items():
    d[city] = sale
s = sorted(d)
for city in s:
    print(city)
```

**Question:** The above code is developed to sort the cities according to their sales volumes from lowest to highest. However, the output is not correct. Please illustrate (1) what the correct output should be, (2) what the error is in the code, (3) how to correct the error by following a Decorate-Sort-Undecorate pattern.

<font color='red'>Edit this cell to answer the question above </font>  
(1) The correct output should be: Tokyo, Hong Kong, Guangzhou, San Francisco
(2) The error is that the code sorts the city names alphabetically instead of sorting them by their sales volumes.  
(3) To fix the error, we should create a list including the city and sales, and then sort the list.


```python
# write the corrected code below:
sales = {'Hong Kong':90, 'Guangzhou':98, 'Tokyo': 88, 'San Francisco': 99}
tmp = []
for city, sale in sales.items():
    tmp.append((sale, city))
tmp.sort()
for sale, city in tmp:
    print(city)

```

**Question 6:** Write the code: (10 points)

```python
#edit this cell to write your code
input_file = input('Enter input file name: ')
max_qty = -1

with open(input_file, 'r') as f:
    for line in f:
        line = line.strip()
        if line == '-1' or not line: # check if it's empty line or -1
            continue
        part = line.split(" ")
        qty = int(part[2])
        origin = part[0]
        destination = part[1]
        if qty > max_qty:
            max_qty = qty
if max_qty != -1:
    print(f'The largest shipping quantity from {origin} to {destination} is {max_qty}.')

```
