```python
def insert_head1(lst, element):
    lst = [element] + lst

def insert_head2(lst, element):
    lst = [element] + lst
    return lst


def insert_head3(lst, element):
    lst.insert(0,element) #Hint: use help() to understand the usage of method insert of a list object
    
    
months=['Mar', 'Apr']

insert_head1(months, 'Feb')
print(months)

months = insert_head2(months, 'Feb')
print(months)

insert_head3(months, 'Jan')
print(months)
```

Answer: because it does not return at the end of the function, so the original list months is not affected.

- (2) For Line 2of the output: How is ``months`` changed? Why?

Answer: The 'insert_head2" returns a new list with 'Feb' inserted at the head, and then assigns it to months, so months is changed.

- (3) For Line 3 of the output: How is ``months`` changed? Why?

Answer: insert function means inserting an element at a specified position in the list, so months is changed by inserting 'Jan' at index 0.

##### Question 5.1.3: Explain how are you going to transform each line of the input file to a list of strings with no leading or trailing white spaces, and how to assign strings from the list to variables that store data of customer_id, customer_name, product_name?

<font color='red'>Edit this cell to answer the question above:</font>  
To transform each line of the input file to a list of strings with no leading or trailing white spaces, I will use the split function with ', ' as the separator. This will split the line into a list of strings. Then I would assign each element of the list to the variables customer_id, customer_name, product_name, and promotion_code respectively by allocating them to the right index of the list.


##### Question 5.1.4: Write the code:

```python
#edit this cell to write your code
file1 = open('sales.in.txt', 'r')
for line in file1:
    line = line.strip()
    if line == '-1':
        break
    customer_info = line.split(', ')
    customer_id = customer_info[0]
    customer_name = customer_info[1]
    product_name = customer_info[2]
    promotion_code = customer_info[3]
    print(f'Dear {customer_name}, Thank you for purchasing {product_name}. You can use the following promotion code to receive a 20% discount in your next purchasing.\n\n{promotion_code}\n\nWish you Merry Christmas and Happy New year!\n\nRegards,\nCompany PolyU\n')

file1.close()
    
```
