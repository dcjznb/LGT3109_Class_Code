#### 4.2. (Event Record) Write a function to print a record of an event in the following format (25 points):
    {Event Name}    
    Location: {Location}                   
    Time: {Time}
    Participants: {Participants}
    
where {Event Name}, {Location}, {Time}, and {Participants} are four parameters of the function.

For example, for an event of my weekly research meeting with students, the record printed is as follows:

    Weekly Research Meeting    
    Location: M505e
    Time: 10:00, Every Friday
    Participants: Professor Xiaoyu Wang and his students

##### (a)

```python
#Edit this cell to define the function and write its body:
def event_generate(Event_name, Location, Time, Participants):
    event = (
        f"{Event_name},\n"
        f"Location:{Location},\n"
        f"Time:{Time},\n"
        f"Participants:{Participants}"
    )
    return event
```

##### (b)

```python
#Edit this cell to write a statement to use the function above to print the event of my weekly research meeting:
event1 = event_generate("Weekly Research Meeting", "M505e", "10:00, Every Friday", "Professor Xiaoyu Wang and his students")
print(event1)
```
