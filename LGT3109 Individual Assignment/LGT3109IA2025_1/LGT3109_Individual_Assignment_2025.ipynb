I would apply the package openpyxl to automate the process of writing data to the Excel file and formatting the worksheet. 

```python
#(b) Edit this cell to present your python code
import openpyxl
from openpyxl.styles import Font, Alignment

def contain_name(line):
    if line == '' or not line[0].isupper():
        return False
    if len(line) > 1 and line[1] in '0123456789':
        return False
    return True

def extract_name(line):
    text = line.strip()
    half_length = len(text) // 2
    first_half = text[:half_length]
    second_half = text[half_length:]
    if first_half == second_half:
        return first_half
    else:
        return text

def contain_code(line):
    if line == '' or line[0] != ' ':
        return False
    if 'T1' in line or 'T2' in line:
        return True
    return False

def extract_code(line):
    text = line.strip()
    parts = text.split()
    return parts[0]

def contain_aisle(line):
    if line == '' or line[0] != ' ':
        return False
    if ' T1 ' in line or ' T2 ' in line:
        return True
    return False

def extract_aisle(line):
    text = line.strip()
    parts = text.split()
    return parts[2]

def contain_tel_eqn(line):
    if line == '':
        return False
    text = line.strip()
    if text.startswith('-') or text.startswith('W1') or text.startswith('E1'):
        if 'airline website' not in text:
            return True
    return False

def extract_tel_eqn(line):
    text = line.strip()
    parts = text.split()
    if len(parts) >= 3:
        if parts[-3] in ['+852', '+800', 'E1', 'W1']:
            return parts[-3] + ' ' + parts[-2] + ' ' + parts[-1]
        else:
            return parts[-2] + ' ' + parts[-1]
    else:
        return ' '.join(parts[1:]) if len(parts) > 1 else text


def create_excel_from_text(txt_file, excel_file):
    print('Reading file: ' + txt_file)
    file = open(txt_file, 'r', encoding='utf-8')
    
    names = []
    codes = []
    aisles = []
    phones = []
    
    for line in file:
        if contain_name(line):
            names.append(extract_name(line))
        elif contain_code(line):
            codes.append(extract_code(line))
            aisles.append(extract_aisle(line))
        elif contain_tel_eqn(line):
            phones.append(extract_tel_eqn(line))
    
    file.close()
    
    print('Data extracted:')
    print('  Names: ' + str(len(names)))
    print('  Codes: ' + str(len(codes)))
    print('  Aisles: ' + str(len(aisles)))
    print('  Phones: ' + str(len(phones)))
    
    min_length = min(len(names), len(codes), len(aisles), len(phones))
    
    if len(names) != len(codes) or len(names) != len(aisles) or len(names) != len(phones):
        print('Warning: Data length mismatch! Using minimum length: ' + str(min_length))
    
    print('Creating Excel file...')
    wb = openpyxl.Workbook()
    ws = wb.active
    ws.title = 'Airline Information'
    
    for i in range(min_length):
        row_num = i + 1
        ws.cell(row=row_num, column=1, value=names[i])
        ws.cell(row=row_num, column=2, value=codes[i])
        ws.cell(row=row_num, column=3, value=aisles[i])
        ws.cell(row=row_num, column=4, value=phones[i])
    
    ws.column_dimensions['A'].width = 30
    ws.column_dimensions['B'].width = 15
    ws.column_dimensions['C'].width = 10
    ws.column_dimensions['D'].width = 35
    
    for row_num in range(1, min_length + 1):
        for col_num in range(1, 5):
            cell = ws.cell(row=row_num, column=col_num)
            cell.alignment = Alignment(horizontal='left', vertical='center')
            cell.font = Font(size=10)
        ws.row_dimensions[row_num].height = 18
    
    wb.save(excel_file)
    print('Excel file created successfully: ' + excel_file)
    print('Total rows written: ' + str(min_length))

txt_filename = input('Enter text file name: ')
excel_filename = 'airline_info_auto.xlsx'
create_excel_from_text(txt_filename, excel_filename)
```
