##### Question 5.1.6: Based on your answers to Questions 5.4 and 5.4, write the code for the problem 5.1:

```python
#edit this cell to write your code
#Hint: (1) input the name of the input file
#      (2) define a file handle for the input file
#      (3) for each line of the file, extract the business unit's name and cargo quantity by using string's find function 
#          and slice operation, and then update the variables that store the largest cargo quantity and its business unit's name

file_name = input('Enter the input file name: ')
input_file = open(file_name, 'r')

max_quantity = -1
max_name = ''

for line in input_file:
    line = line.strip()
    if line == '-1' or line == '':
        break
    p = line.find(' ')
    name = line[:p]
    quantity = int(line[p+1:])
    if quantity > max_quantity:
        max_quantity = quantity
        max_name = name
input_file.close()
print(f'The largest cargo quantity is {max_quantity}, submitted by {max_name}.')
```
